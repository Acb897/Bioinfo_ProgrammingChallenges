# Manipulating data frames ("tibbles")

Data tables such as the gapminder datasets are represented as objects called "tibbles" in R.  

Manipulating tibbles is achieved by calling functions from other libraries.  One of these is called ["dplyr"](https://cran.r-project.org/web/packages/dplyr/)

    dplyr: A Grammar of Data Manipulation

    A fast, consistent tool for working with data frame like objects, both in memory and out of memory.

We will look at only three kinds of manipulation:  **Filtering, Arraging, and Mutating**



# Filtering in dplyr

Filtering means to eliminate rows of data that are not interesting.  This is achieved by selecting a column (based on the name of the column - the "header" of that column) and the values that are of interest.  

We will now select only the gapminder data from the year 2007.  In dplyr, there is a concept of "pipes" very similar to pipes in Unix.  To utilize dplyr filters, you "pipe" the dataset through the filter.  The "pipe" symbol is:

    %>%
    
The dplyr command is

    filter
    
So the instruction looks like this:

    gapminder %>% filter(year == 2007)   # note the "==", just like in Ruby
    



```r
library(dplyr)      # dplyr provides functions to filter, arrange, and mutate datasets

gapminder %>% filter(year == 2002)
```

# To assign a result to a new variable, use the <- operator

You may want to preserve certain "states" of the dataset.  To do this, you can assign the output to a new variable.


```r
library(gapminder)
library(dplyr)

gm_2007 <- gapminder %>% filter(year == 2007)

head(gm_2007)
```

# Arrange in dplyr

You can use the "arrange" function to put data in order.  Again this is a piped function.  A typical command to arrange the gapminder dataset by popuation, from highest to lowest, would be:

    gapminder %>% arrange(desc(pop))


```r
library(gapminder)
library(dplyr)

gapminder %>% arrange(desc(pop))
```

# pipes work like pipes

The concept of "pipes" suggests that you can pass data from one to another.  You can!  To both Filter and Arrange, you do this:

    gapminder %>% arrange(desc(pop)) %>% filter(year == 2007)
    


```r
library(gapminder)
library(dplyr)

gapminder %>% arrange(desc(pop)) %>% filter(year == 2007)
```

# Mutate function of dplyr

In addition to filter and arrange, you sometimes want to manipulate the data values before analysing or visualizing them.  This can be achieved using the mutate function.  Like other dplyr functions, it is a pipe.  For example, if we wanted to view the population in terms of millions of people, we could do this:

     gapminder %>% mutate(pop = pop/1000000)

```r
library(gapminder)
library(dplyr)

inmillions <- gapminder %>% mutate(pop = pop/1000000)
head(inmillions, n = 5L)   # n = number of lines to show

asnewcolumn <- gapminder %>% mutate(pop_in_mil = pop/1000000)  # creates a new column
head(asnewcolumn, n = 5L)
```

# Putting it all together

You can use pipes to do all of these things in one operation:



```r
library(gapminder)
library(dplyr)

beautiful_data <- gapminder %>% mutate(pop_in_mil = pop/1000000) %>% filter(year == 2007) %>% arrange(desc(pop))

head(beautiful_data, n = 5L)

```

## Read and write data

The full command for reading a file is:

    read.csv(filename, headers?, separator)
    
where "filename" is either the full path to the file, or just the name (which will default to be in the working directory).  For example, we will now load the "secretome" file that we created from UniProt into a tibble called 'secretome':

    secretome <- read.csv('secretome.csv', TRUE, ",")
    
This means that the file is called "secretome.csv", it DOES have a header line, and the elements are comma-separated.
    

```r
secretome <- read.csv('secretome.csv', TRUE, ",")
head(secretome)
secretome_arr <- secretome %>% arrange(desc(mass))
head(secretome_arr)
```

<pre>


</pre>
Writing data is almost the same, but has one additional parameter - that is, should we write-out the first column, or not?  Often, the first column is just an index (line 1, line 2, line 3, etc.), so if it is, then you don't need to add another row name.  It is entirely up to you.  The commands, of course, respect the current working directory (setwd()), and the command structures are:

    write.csv(tibble, "filename.csv", row.names = F)  # first column gone!
    write.csv(tibble, "filename.csv", row.names = T)  # blank column name added!

Watch what happens when we use these commands:


```r
write.csv(secretome_arr, "secretome_arranged_false.csv", row.names = F)
write.csv(secretome_arr, "secretome_arranged_true.csv", row.names = T)

withnames = read.csv("secretome_arranged_true.csv", TRUE, ",")
withoutnames = read.csv("secretome_arranged_false.csv", TRUE, ",")

head(withnames)   # the column is called "X", and it is automatically indexed.
head(withoutnames)
```

```r
library(ggplot2)  # a library of graphics
secretome <- read.csv('secretome.csv', TRUE, ",")
ggplot(secretome, aes_string(quote(mass), quote(length))  )   + geom_point()
```

# Manipulate our secretome data

All of our data points are "bunched up" at the bottom left corner of the graph.  It might be better to use a log-scale.  The "aesthetics" of the x and y axis are, by default, an aesthetic called "scale_x_continuous" and "scale_y_continuous".  You can change these.

ggplot allows you to ["add aesthetics" to a graph](http://ggplot2.tidyverse.org/reference/scale_continuous.html).  You do this by putting the graphic into a variable,and then adding new aesthetics to it (literally!).  Watch this:



```r
# size is the size of the point, stroke is the width of the outline of the point , fill is the internal color
# shapes are described on this page: https://cran.r-project.org/web/packages/ggplot2/vignettes/ggplot2-specs.html

mygraph <- ggplot(secretome, aes(x = mass, y = length)  ) + geom_point( size = 2, stroke = 1, shape = 23, fill="red")
mygraph <- mygraph + scale_x_log10()  # change the x axis aesthetic to "scale_x_log10"
mygraph
mygraph <- mygraph + scale_y_log10()   # change the y axis aesthetic to "scale_y_log10"
mygraph
```

#  Now, where are the secreted data?

To visualize the third dimension of our dataset - "secreted" or "not secreted" - we will need to distinguish the points that are associated with each of these factors.  This is also done usng "aesthetics" - in this case, the aesthetics of our geom_point() objects.  There is a good [tutorial about ggplot2 aesthetics](https://cran.r-project.org/web/packages/ggplot2/vignettes/ggplot2-specs.html) that you can use to learn a bit more about aesthetics for plotting.

To solve this problem, we need to consider each of these column values as a "factor" - like "factors" in an experiment. Like Integers and Floats, a "factor" is a specific TYPE of data in R!  Let's see what kinds of data we have in our secretome.  The command is <code>str(dataset)</code>:

The report below shows us that we have mass and length interpreted as "int" (integer), but protein_status is interpreted as a "Factor" data-type:

    $ protein_status: Factor w/ 2 levels "not_secreted",..: 2 2 2 2 2 2 2 2 2 2 ...
    


```r
str(secretome)
```

## R is very smart, and it already did "the right thing"!

So R has already interpreted the values in that column as a 'Factor'.  Perfect!  Factors are automatically interpreted by ggplot, and can be used to assign things like color or size.  In fact, ggplot can do this for you automatically!  Look at the code below:

```r


mygraph <- ggplot(secretome, aes(x = mass, y = length)  ) +   # note that the "+" symbol must appear on this line
                            geom_point( size = 2, stroke = 1, shape = 23,  # everything to here is identical to before
                                      aes(fill = protein_status) # create a new "aesthetic" 
                                    )         # that tells the 'fill' parameter to look at the Factor prot_stat
    
mygraph <- mygraph + scale_x_log10()  # change the x axis aesthetic to "scale_x_log10"
mygraph <- mygraph + scale_y_log10()   # change the y axis aesthetic to "scale_y_log10"
mygraph
```

## This is still not what we want...

We know that there are about 30 "secreted" proteins in our dataset, but we can only see 2 of them!  The rest are covered by the not_secreted points.  How do we bring them to the front?  

2 solutions.  First solution, is to plot the "secreted" last, and we can do that by ordering the dataset:


```r
# the code below allows you to swap the two colors


secretome <-secretome %>% arrange(protein_status)  # we are ordering the protein_status so that "secreted" come last

mygraph <- ggplot(secretome, aes(x = mass, y = length)  ) + 
           geom_point( aes(fill = protein_status), size = 2, stroke = .4, shape = 23)
mygraph <- mygraph + scale_x_log10()  # change the x axis aesthetic to "scale_x_log10"
mygraph <- mygraph + scale_y_log10()   # change the y axis aesthetic to "scale_y_log10"
mygraph

# if you don't like the colors that are automatically selected
# look at the instructions here:  http://ggplot2.tidyverse.org/reference/scale_manual.html

colrs <- c("secreted" = "red", "not_secreted" = "blue")  # create a hash
mygraph_redblue <- mygraph + scale_fill_manual(values = colrs)
mygraph_redblue
```

# Just FYI

You can play with the code below to explore other kinds of visualizations and aesthetics.  It also demonstrates how you create lists and hashes in R.


```r
lty <- c("blank", "solid", "dashed", "dotted", "dotdash", 
     "longdash","twodash")  # this is similar to a Ruby list

linetypes <- data.frame(   # data frames are tibble-like things, so they can be used as data sources
  y = seq_along(lty),
  lty = lty
) 
#linetypes   # this is a data frame (like a tibble) - print it to see what it contains.  
            # "seq_along" is the function that created the indexes for each element of lty


ggplot(linetypes, aes(0, y)) +   # takes each value of "y" from linetypes, iteratively...
  geom_segment(   aes(xend = 5, yend = y, linetype = lty)   ) + 
  scale_linetype_identity(guide = "legend") +    # the figure legend!
  geom_text(aes(label = lty), hjust = 0, nudge_y = 0.1) +
  scale_x_continuous(5, breaks = 1:5) +   # these are the white vertical lines
  scale_y_continuous(NULL, breaks = NULL)
```

```r
library(ggplot2)

jobsAFAM1 <- data.frame(
  data_date = runif(5,1,100),
  Percent.Change = runif(5,1,100)
)

jobsAFAM2 <- data.frame(
  data_date = runif(5,1,100),
  Percent.Change = runif(5,1,100)
)

ggplot() + 
  geom_line(data = jobsAFAM1, aes(x = data_date, y = Percent.Change), color = "red") +
  geom_line(data = jobsAFAM2, aes(x = data_date, y = Percent.Change), color = "blue") +
  xlab('data_date') +
  ylab('percent.change')
```
